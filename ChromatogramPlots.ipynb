```python
#Code written by Thomas Dixon
import numpy as np
import plotly.express as px
import pandas as pd
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import plotly.colors as pc
```

```python
#Import all data

data1 = pd.read_excel(r"Optimisation1AllData.xls",
                    header = None)
data2 = pd.read_excel(r"Optimisation2AllData.xls",
                    header = None)
data3 = pd.read_excel(r"Optimisation3AllData.xls",
                    header = None)

variables1 = pd.read_excel(r"Optimisation1.xlsx")
variables2 = pd.read_excel(r"Optimisation2.xlsx")
variables3 = pd.read_excel(r"Optimisation3.xlsx")

allVars = [variables1, variables2, variables3]
allVars = pd.concat(allVars)

minInd1 = variables1.WeightRespcmtmdHPLC018.iloc[0:20].idxmin()
print(minInd1)
print(variables1.iloc[minInd1,:])
minInd2 = variables2.WeightRespcmtmdHPLC018.iloc[0:20].idxmin()
print(minInd2)
print(variables2.iloc[minInd2,:])
minInd3 = variables3.WeightRespcmtmdHPLC018.iloc[0:20].idxmin()
print(minInd3)
print(variables3.iloc[minInd3,:])

data1 = data1.iloc[:,[minInd1*2, (minInd1*2)+1]]
data2 = data2.iloc[:,[minInd2*2, (minInd2*2)+1]]
data3 = data3.iloc[:,[minInd3*2, (minInd3*2)+1]]

chrome1 = pd.read_excel(r"Optimisation4Exp17.xlsx")
chrome2 = pd.read_excel(r"Optimisation7Exp21.xlsx")

```

```python
#Figure 3. The first chromatograms from Optimisations 1 (top), 2 (middle) and 3 (bottom) 
#that satisfied the design space criteria (N = 7, RsCrit  2 and tRL  6 minutes), 
#representing experiments 12, 19 and 9 respectively.

config = {
  'toImageButtonOptions': {
    'format': 'png', # one of png, svg, jpeg, webp
    'filename': 'custom_image',
    'height': 800/2,
    'width': 1000/2,
    'scale':6 # Multiply title/legend/axis/canvas sizes by this factor
  }
}

time3 = data3.iloc[:,0]
inten3 = data3.iloc[:,1] + 2e5
trace3 = go.Scatter(x=time3,
                    y=inten3,
                    mode='lines',
                    line = dict( color = "rgb(255,176,0)", width= 2.5),
                    name= "Optimisation 3")

time2 = data2.iloc[:,0]
inten2 = data2.iloc[:,1] + data3.iloc[:,1].max() + 1e4 + 2e5
trace2 = go.Scatter(x=time2,
                    y=inten2,
                    mode='lines',
                    line = dict( color = "rgb(254,70,0)", width= 2.5),
                    name= "Optimisation 2")

time1 = data1.iloc[:,0]
inten1 = data1.iloc[:,1] + data2.iloc[:,1].max() + data3.iloc[:,1].max() + 1e4 + 2e5
trace1 = go.Scatter(x=time1,
                    y=inten1,
                    mode='lines',
                    line = dict( color = "rgb(220,38,127)", width= 2.5),
                    name= "Optimisation 1")


fig = make_subplots(specs=[[{"secondary_y": True}]])
fig.add_trace(trace1)
fig.add_trace(trace2)
fig.add_trace(trace3)

fig.update_yaxes(visible=True, showticklabels=False,showgrid=False)
fig.update_xaxes(showgrid=False)   

fig.update_layout(autosize=False, 
                  width=500, 
                  height=400, 
                  showlegend=False, 
                  plot_bgcolor="white",template='plotly_white')

fig['layout']['xaxis']['title']='Time / minutes'
fig['layout']['yaxis']['title']='Intensity'

fig.update_xaxes(range = [0, 6],showgrid=False, gridwidth=2, tickfont=dict(color='black',size=15),ticks='outside',tick0=0, 
                 dtick = 1, tickwidth=4, title = dict(font = dict(size=13)), zerolinecolor = "black")
fig.update_yaxes(range = [-1e5,0.9e7],showgrid=False, gridwidth=2, tickfont=dict(color='black',size=15),ticks='outside',tick0=0, 
                 dtick = 1, tickwidth=4, title = dict(font = dict(size=13)), zerolinecolor = "black", showticklabels  = False)

fig.show()

import plotly.io as pio
pio.write_image(fig, 'cmtmdHPLC018-020BestChromes.jpg', scale=4, width=1000/2, height=800/2)
```

```python
#Figure 7. Optimal chromatograms for Mixture A (top) and Mixture B (bottom) using TS-EMO, 
#representing Optimisation 4 experiment 17 and Optimisation 7 experiment 30 re-spectively.

time2 = chrome2.iloc[:,0]
inten2 = chrome2.iloc[:,1] 
trace2 = go.Scatter(x=time2,
                    y=inten2,
                    mode='lines',
                    line = dict( color = "black", width= 2.5),
                   name= "Eleven component seperation")

time1 = chrome1.iloc[:,0]
inten1 = chrome1.iloc[:,1] + chrome2.iloc[:,1].max()/2.1
trace1 = go.Scatter(x=time1,
                    y=inten1 + 140,
                    mode='lines',
                    line = dict( color = "rgb(220,38,127)", width= 2.5),
                   name= "Seven component seperation")


fig = make_subplots(specs=[[{"secondary_y": True}]])
fig.add_trace(trace1)
fig.add_trace(trace2)

fig.update_yaxes(visible=True, showticklabels=False,showgrid=False)
fig.update_xaxes(showgrid=False)   

fig.update_layout(autosize=False, 
                  width=500, 
                  height=400, 
                  showlegend=False, 
                  plot_bgcolor="white",template='plotly_white')


fig['layout']['xaxis']['title']='Time / minutes'
fig['layout']['yaxis']['title']='Intensity'

fig.update_xaxes(range = [0, 6],showgrid=False, gridwidth=2, tickfont=dict(color='black',size=15),ticks='outside',tick0=0, 
                 dtick = 1, tickwidth=4, title = dict(font = dict(size=13)), zerolinecolor = "black")
fig.update_yaxes(range = [0,400],showgrid=False, gridwidth=2, tickfont=dict(color='black',size=15),ticks='outside',tick0=0, 
                 dtick = 1000, tickwidth=4, title = dict(font = dict(size=13)), zerolinecolor = "black", showticklabels  = False)

fig.show()

import plotly.io as pio
pio.write_image(fig, 'BestChromesOverall.jpg', scale=4, width=1000/2, height=800/2)
```
